```python
import pandas as pd
from scipy.stats import mannwhitneyu
```

```python
df = pd.read_excel('Данные для тестового задания.xlsx', sheet_name=1)
df.head()
```

```python
experiment = df['experiment_num'].unique()
experiment
```

```python
for i in experiment:
    data = df[df['experiment_num'] == i]

    control_group_rev = data[data['experiment_group'] == 'control']['revenue']
    test_group_rev = data[data['experiment_group'] == 'test']['revenue']

    stat, p_value = mannwhitneyu(control_group_rev, test_group_rev, alternative='two-sided')

    arpu_control_group = control_group_rev.mean()
    arpu_test_group = test_group_rev.mean()

    print(f'AB test for experiment_num: {i}')
    print(f'ARPU control: {arpu_control_group:.2f}')
    print(f'ARPU test: {arpu_test_group:.2f}')
    print(f'p-value: {p_value:.4f}')
    print('='*50)
    
```
